Using https://s3-us-west-2.amazonaws.com/dsci/6002/data/email.csv, classify emails as `spam` or not with `line_breaks` and `attach`. Interpret the model coefficients for the two predictors.

```python
%pylab inline

import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('https://s3-us-west-2.amazonaws.com/dsci/6002/data/email.csv')
df.columns
df.head()
```

```python
logit_model1 = smf.logit("spam ~ line_breaks + attach", data=df).fit()
logit_model1.summary()
```

attach is positivly correlated with spam, while line_breaks is negatively correlated.  So higher attach leads to higher probability of spam, and symmetricly for line_breaks.

more line breaks makes it less likely it's spam


the coefficients are the log odds.

```python
# logit function
def f(t):
    return exp(t) / (1 + exp(t))
```
